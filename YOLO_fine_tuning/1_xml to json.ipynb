```python
import os
import xml.etree.ElementTree as ET
import json

# XML 파일이 있는 폴더 경로
folder_path = "C:/Users/sunxo/Study/BIS/data/xml/절도"

# JSON 파일을 저장할 폴더 경로
json_folder_path = "C:/Users/sunxo/Study/BIS/data/JSON/절도"

# JSON 파일이 완료되었는지를 카운트하기 위한 변수 초기화
json_file_count = 0

# 이전에 저장한 GENDER 값을 저장하는 변수 초기화
prev_gender = None

# 폴더 내의 모든 XML 파일에 대해 반복
for filename in os.listdir(folder_path):
    if filename.endswith(".xml"):
        xml_path = os.path.join(folder_path, filename)

        with open(xml_path) as f:
            tree = ET.parse(f)
            root = tree.getroot()

        data = []

        # track 요소 안의 box 요소들을 찾아 반복합니다.
        for box_element in root.findall('.//box'):
            frame = box_element.get('frame')
            gender_elem = box_element.find('.//attribute[@name="gender"]')

            # GENDER 항목이 없으면 현재 값을 저장하고, 없으면 이전에 저장한 값을 사용합니다.
            if gender_elem is not None:
                prev_gender = gender_elem.text

            xtl = box_element.get('xtl')
            ytl = box_element.get('ytl')
            xbr = box_element.get('xbr')
            ybr = box_element.get('ybr')

            # GENDER 항목과 이전에 저장한 GENDER 값이 모두 없으면 해당 프레임을 무시하고 다음으로 넘어갑니다.
            if gender_elem is None and prev_gender is None:
                continue

            box_data = {
                "frame": frame,
                "gender": prev_gender,
                "xtl": xtl,
                "ytl": ytl,
                "xbr": xbr,
                "ybr": ybr
            }

            data.append(box_data)
            prev_gender = None

        # 파일 이름 추출
        file_name_without_extension = os.path.splitext(filename)[0]

        # JSON 파일을 저장할 폴더 경로
        json_file_path = os.path.join(json_folder_path, f"{file_name_without_extension}.json")

        # 현재 파일의 JSON 파일로 저장
        with open(json_file_path, 'w') as json_file:
            json.dump(data, json_file, indent=2)

        print(f"JSON 파일 {json_file_path} 이 생성되었습니다.")

        # JSON 파일이 완료되었으므로 카운트 증가
        json_file_count += 1

# JSON 파일이 N개 완료되었습니다를 출력
print(f"JSON 파일 {json_file_count}개가 완료되었습니다.")
```
